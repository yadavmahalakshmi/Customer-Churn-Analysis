```python
#Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Churn count plot
sns.countplot(x='Churn',data=df)
plt.title('Churn Count Plot')
plt.show()
```

```python
#Churn by Contract
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title('Churn by Contract')
plt.show()
```

```python
#Monthly Charges Distribution
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges Distribution')
plt.show()
```

```python
#Tenure vs Churn
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tenure VS Churn')
plt.show()
```

```python
df['AvgCharges']=df['TotalCharges'] / df['tenure'].replace(0,1)
```

```python
df['AvgCharges']
```

```python
#Correlation HeatMap
numeric_df=df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Churn Percentage Visualization
churn_percent = df['Churn'].value_counts(normalize=True) * 100
churn_percent.plot(kind='bar')
plt.title('Churn Percentage')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
#Churn Rate By Category
contract_churn = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
contract_churn.plot(kind='bar', stacked=True)
plt.title('Churn Rate by Contract Type (%)')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
#Tenure Group Analysis
df['TenureGroup'] = pd.cut(df['tenure'],
                           bins=[0,12,24,48,72],
                           labels=['0-1yr', '1-2yr', '2-4yr', '4-6yr'])

tenure_churn = pd.crosstab(df['TenureGroup'], df['Churn'], normalize='index') * 100
tenure_churn.plot(kind='bar', stacked=True)
plt.title('Churn Rate by Tenure Group')
plt.ylabel('Percentage')
plt.show()
```

```python
#Monthly Charges Distribution
sns.histplot(data=df, x='MonthlyCharges', hue='Churn', kde=True)
plt.title('Monthly Charges Distribution by Churn')
plt.show()
```

```python
#KPI Summary
print('Total Customers: ',len(df))
print('Churn Rate (%): ',round(df['Churn'].mean()*100,2))
print('Average Monthly Charges: ',round(df['MonthlyCharges'].mean(),2))
```

```python
#Churn by Payment Method %
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100
payment_churn.plot(kind='bar', stacked=True)
plt.title('Churn by Payment Method (%)')
plt.show()
```

```python
#Insights
1.Customers with higher monthly charges exhibit a significantly higher churn rate, indicating strong price sensitivity among users.
2.Customers on month-to-month contracts show the highest churn rate, whereas long-term contracts (one-year and two-year) demonstrate better customer retention.
3.New customers (low tenure) are more likely to churn, suggesting that early-stage customer experience plays a critical role in retention.
4.High-paying customers with short tenure represent a high-risk revenue segment, as they contribute more revenue but churn quickly.
5.Customers with specific services (like fiber internet) tend to show higher churn variability, indicating possible service quality or pricing concerns.
```

```python
#Business Recommendations
1.Introduce flexible pricing plans or discounts for high monthly charge customers to reduce churn caused by cost sensitivity.
2.Encourage customers to shift from month-to-month plans to long-term contracts by offering incentives such as discounts or bundled services.
3.Focus on improving the first 3–6 months customer experience, including better onboarding, support, and engagement strategies.
4.Identify high-risk customers (low tenure + high charges) and target them with personalized retention campaigns.
5.Analyze and improve services with higher churn rates, especially internet services, to enhance customer satisfaction.
```
